 ## Zomato Data Set Analysis and Visualization

### Importing Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

### Read CSV

```python
df=pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.drop(['url','address','phone','reviews_list','menu_item','dish_liked'],axis=1)
df.head()
```

```python
df.info()
```

## Dropping Duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

## Cleaning Rate Coloumn

```python
df['rate'].unique()
```

```python
def cleanrate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(cleanrate)
df['rate'].head()
```

```python
df.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

#### Dropping Null Values in other columns since it's not making much difference

```python
df.dropna(inplace=True)
df.head()
```

## Analysing Columns

```python
df.rename(columns={'approx_cost(for two people)':'Cost_2','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

### Since both contain locations, let's consider one column that's relevant

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df['Cost_2'].unique()
```

### There's a seperator in between the numbers making the integer a string, so we have to convert it 

```python
def comma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost_2']=df['Cost_2'].apply(comma)
df['Cost_2'].unique()
```

## Clean 'Rest' Column

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

### Since we can see that some of the rest types have less values, we can cluster them into "Other"

```python
rest_less=rest_types[rest_types<1000]
rest_less
```

```python
def resto(value):
    if (value in rest_less):
        return 'others'
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(resto)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
loca_less=location[location<300]
loca_less
```

```python
def loca(value):
    if (value in loca_less):
        return 'Others'
    else:
        return value
df['location']=df['location'].apply(loca)
df['location'].value_counts()
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisi=cuisines[cuisines<100]

def cu(value):
    if (value in cuisi):
        return "others"
    else:
        return value
df['cuisines']=df['cuisines'].apply(cu)
df['cuisines'].value_counts()
```

```python
df. head()
```

```python
df['Type'].value_counts()
```

## Visualization

### Count Plot on Locations

```python
plt.figure(figsize=(16,10))
ax=sns.countplot(data=df,x='location')
plt.xticks(rotation=90)
plt.show()
```

### A take away/insight = Maybe we could open a restaurant in Old Airport road or St Marks Road, since there's few restaurants opened up there

### Visualizing How Many Restos have Online order facilities?

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df,x='online_order',palette='inferno')
```

### Visualizing Book Table Facilities available

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df,x='book_table',palette='rainbow')
```

### Is there any correlation between booking tables and ratings?

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=df,x='book_table',y='rate')
```

### Business Takeway : Keep in mind- Placing a book table facility can lead to increase in average ratings

### Number of Votes Location Wise
